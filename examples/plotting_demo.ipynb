## pycasaxps plotting demo
This walks through the basic functionality of the module with some sample data. This module is intended for use with ASCII files exported from CasaXPS after creating peak fits.

```python
"Import required packages"
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#this imports the casa_data class from the casa_plot module
from pycasaxps import casa_data
```

## Quick overview
This shows the full code for loading data, creating a plot, and saving it as an image.

```python
"Making a nice plot"
au = casa_data('ascii_Au4f.txt') #load the data
au.rename(components=['4f7/2','4f5/2','Ba4d5/2']) #rename the components
fig, ax = au.plot(cyclw=5,title='Au 4f',yticks=True) #plot the data
ax.set_ylim([0,25000]) #rescale the y axis to fit the labels
fig.tight_layout() #correct the plot boundaries
plt.savefig('Au_4f_plot.png',dpi=400) #save the figure as an image
```

## Step-by-step
Calling the `casa_data` class creates an instance of the class based on the file specified. This instance will read an ASCII data file exported from CasaXPS, and store it in a dataframe. (To export your data from CasaXPS, click on the right side of the screen, click the "Save Tab ASCII" button, write a filename, and choose the "Rows of Tables" option when prompted.)

```python
au = casa_data('ascii_Au4f.txt') #read the Au_4f.txt file (exported from CasaXPS)
au.data #display the data read from the file
```

The "Cycle 0:..." column is the measured data. The "Au_4f_sect" columns are the CasaXPS quantification components (peak fits) for the Au4f7/2, and Au4f5/2, and Ba4d5/2 peaks. "Background" is the background, and "Envelope" is the total fit. We can give these columns less ugly names with the `rename` function. The `components` argument takes a list of component names (in the same order that the components appear in the dataframe). The "Cycle" column will be renamed automatically.

```python
"rename the dataframe columns to more convenient names"
au.rename(components=['Au4f7/2','Au4f5/2','Ba4d5/2']) #assign names for the components
au.data #display the modified dataframe
```

Now we're ready to plot the data. The `plot` function will plot the measured data, the components, the background, and the overall fit. It will also automatically label the peaks based on the component names.

```python
au.plot() #plot the data with default settings
```

As you can see, `plot` does not adjust the axes to make room for the peak labels; this must be done manually. It also defaults to displaying the y-axis as an arbitrary intensity, without ticks or values shown. This can be changed by setting the `yticks` option to `True`.

```python
au.plot(yticks=True) #plot with y tick labels
plt.ylim([0,25000]) #increase the y scale to make room for the peak labels
```

We can also add a title to indicate the core level for the spectrum, and increase the linewidth of the cycle (measured data) so that it is visible behind the fit lines.

```python
#plot with y tick labels, title, and thicker cycle linewidth
au.plot(yticks=True,title='Au 4f',cyclw=5) 
plt.ylim([0,25000]) #increase the y scale to make room for the peak labels
```

The `peaks` attribute also provides a convenient way to check peak positions. It returns a dictionary that shows the x (BE) and y (intensity) coordinates for each fitted peak.

```python
au.peaks #show all peaks
```

```python
au.peaks['Au4f7/2'] #show the coordinates for the Au4f7/2 peak
```

The `casa_data` class can also handle exported data with no fits applied, although the plots are less interesting. Here's an example using a scan of the N 1s region that didn't show any peaks.

```python
n1s = casa_data('ascii_N1s.txt')
n1s.plot(title='N 1s',yticks=True)
```

The x-axis ticks can be adjusted using the `xmin` and `xint` options. `xmin` sets the location of the first (lowest BE) x tick, and `xint` sets the interval between ticks.

```python
n1s.plot(title='N 1s',yticks=True,xmin=396,xint=4)
```
